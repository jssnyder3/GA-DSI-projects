```python
# Create a histogram by creating a list comprehension + a function to turn the data into inches

def cleanup(x):
    if pd.isnull(x):
        return x
    elif 'Under' in x:
        return 78
    elif 'above' in x:
        return 60
    else:
        feet = int(x.split("'")[0]) * 12
        inches = int(x.split("'")[1].replace('"',''))
        return feet + inches

fte['Height'] = [cleanup(x) for x in fte['Height']]

fte['Height'].hist()
```

```python
education = {}
education = {key:0 for key in fte['Education'].unique()}

education
```

```python
def counter(x):
    education[x] = education[x] + 1

for x in fte['Education']:
    counter(x)

education
```
