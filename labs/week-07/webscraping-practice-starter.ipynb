# Scraping a webpage

Here's a link to the Box Office Mojo page for Jurrasic Park. See if you can scrape the following information:
- Title
- Domestic Total Gross
- Distributor
- MPAA rating

```python
url = 'http://www.boxofficemojo.com/movies/?id=jurassicpark.htm'
```

```python
r = requests.get(url)
```

```python
r.content
```

```python
r_soup = BeautifulSoup(r.content, "lxml")
r_soup.prettify()
```

```python
title = r_soup.find('title').text
title = title.split('(')
title = title[0].strip(' ')
print (title)
```

```python
dist = r_soup.find(text=re.compile("Distributor"))
print (dist.findNextSibling().text)
```

```python
dom_gross = r_soup.find(text=re.compile("Domestic Total Gross"))
print (dom_gross.findNextSibling().text)
```

```python
mpaa = r_soup.find(text=re.compile("MPAA"))
print (mpaa.findNextSibling().text)
```

```python
def movie_info(link):
    r = requests.get(link)
    r_soup = BeautifulSoup(r.content, "lxml")
    title = r_soup.find('title').text
    title = title.split('(')
    title = title[0].strip(' ')
    
    dist = r_soup.find(text=re.compile("Distributor"))
    distributor = dist.findNextSibling().text
    
    dom_gross = r_soup.find(text=re.compile("Domestic Total Gross"))
    dom_gross_total = dom_gross.findNextSibling().text
    
    mpaa = r_soup.find(text=re.compile("MPAA"))
    mpaa_rating = mpaa.findNextSibling().text
    
    print (("Movie title: {}").format(title))
    print (("Distributor: {}").format(distributor))
    print (("Domestic Gross: {}").format(dom_gross_total))
    print (("MPAA rating: {}").format(mpaa_rating))
    
    return title, distributor, dom_gross_total, mpaa_rating
```

# Scraping a few webpages

Let's say we want to get that same information for several movies at once. Here's a link to the top 100 movies from 1993. Use this table to generate a list of URLs from each movie. 



```python
top_100_url = 'http://www.boxofficemojo.com/yearly/chart/?yr=1993&p=.htm'
top = requests.get(top_100_url)
text = top.text
```

```python
msoup = BeautifulSoup(text, "lxml")
```

```python
for i in msoup.findAll('a'):
    print (i.text)
    print (i.get('href'))
```

```python
msoup.a
```

```python
urls = []

for item in msoup.findAll('a', href=re.compile("/movies/")):
    end_url = item['href']
    url = 'http://www.boxofficemojo.com' + end_url
    urls.append(url)
url_list = urls[1:]
url_list
```
