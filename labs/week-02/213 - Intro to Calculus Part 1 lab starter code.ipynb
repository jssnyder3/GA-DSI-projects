*** Exercise 1: ***
Plot the following functions in a 2x2 block using lambda expressions, pyplot.subplots and figure.subplots_adjust (only use the matplotlib libary). Determine the appropriate plotting range and determine the image of the Function. Label each graph appropriatly including ledgend, x-labels, y-labels and title. Which one of the functions is a one-to-one map?
1. $F_1(x) = Exp(-0.5*(x-2)^2)$
2. $F_2(x) = x^2 * Exp(-0.8*(x+2)^2)$
3. $F_3(x) = F_2(x)*F_1(x)$
4. $F_4(x) = \frac{1}{1+Exp(-x)}$

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

X = np.arange(-10,11,0.1)
F_of_X1 = lambda x: np.array(np.exp((-0.5*(x-2)**2)))
F_of_X2 = lambda x: x**2 * np.exp(-0.8*(x+2)**2)
F_of_X3 = lambda x: F_of_X1(x) * F_of_X2(x)
F_of_X4 = lambda x: 1/(1+np.exp(-x))
#FN_of_X =  lambda N: np.array([sum([1/float(np.math.factorial(k))*pow(x,k) for k in range(N+1)]) for x in X])
Functions = [F_of_X1, F_of_X2, F_of_X3, F_of_X4]
Functions

#Create one figure with four graphs
fig, ax = plt.subplots(2,2)   #Creates figure and axis, simultaneously
ax = ax.reshape(4,)
func_lists = zip(Functions, ax)
count = 1
for element in func_lists:
    element[1].plot(X,element[0](X), label = 'F_of_X{}'.format(count))
    element[1].legend(loc = 'best', fontsize = 'small')
    count += 1
    
```

```python
X = np.arange(-10,11,0.1)
F_of_X1 = lambda x: np.array(np.exp((-0.5*(x-2)**2)))
F_of_X2 = lambda x: x**2 * np.exp(-0.8*(x+2)**2)
F_of_X3 = lambda x: F_of_X1(x) * F_of_X2(x)
F_of_X4 = lambda x: 1/(1+np.exp(-x))
#FN_of_X =  lambda N: np.array([sum([1/float(np.math.factorial(k))*pow(x,k) for k in range(N+1)]) for x in X])
Functions = [F_of_X1, F_of_X2, F_of_X3, F_of_X4]
Functions

#Create four figures with four graphs
fig, axes = plt.subplots(2,2)
fig.subplots_adjust(wspace = 0.3, hspace = 0.5)
axes[0,0].plot(X, F_of_X1(X), label = 'F1')
axes[0,1].plot(X, F_of_X2(X), label = 'F2')
axes[1,0].plot(X, F_of_X3(X), label = 'F3')
axes[1,1].plot(X, F_of_X4(X), label = 'F4')
for ee in axes:
    for i in ee:
        i.legend(loc = 'best', fontsize = 'small')
        i.set_xlabel('x')
        i.set_title('Function: F{}'.format(i))
```

*** Exercise 2: ***
Plot the following functions and their respective contour plots in a 2x2 block. Determine the appropriate plotting range and determine the image of the Function. Label each graph appropriatly including ledgend, x-labels, y-labels and title.
1. $G_1(x,y) = Exp(-0.01*(x^3+y^3))$
2. $G_2(x,y) = Log(x^2+2*y^2)*x^2$


```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

G_of_XY1 = lambda x,y: np.exp(-0.01*(x**3 + y**3))
G_of_XY2 = lambda x,y: np.log(x**2 +2*y**2)*x**2

from mpl_toolkits.mplot3d import Axes3D
X = np.linspace(-2,2, 100)
Y = np.linspace(-2,2, 100)
X, Y = np.meshgrid(X,Y)
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1, projection = '3d')
ax1.plot_surface(X, Y, G_of_XY1(X,Y))
ax1.set_title('G1 - 3D Plot')
ax2 = fig.add_subplot(2,2,2)
ax2.contour(X, Y, G_of_XY1(X,Y))
ax2.set_title('G_of_XY2 - Contour Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

ax3 = fig.add_subplot(2,2,3, projection = '3d')
ax3.plot_surface(X, Y, G_of_XY2(X,Y))
ax3.set_title('G2 - 3D Plot')
ax4 = fig.add_subplot(2,2,4)
ax4.contour(X, Y, G_of_XY2(X,Y))
ax4.set_title('G_of_XY2 - Contour Plot')
ax4.set_xlabel('x')
ax4.set_ylabel('y')

fig = plt.figure(figsize=(10,10))
```
