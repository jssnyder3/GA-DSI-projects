```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def RandomWalkLine(R, L):
    Walk = np.random.uniform(low = -1, high = 1, size =(R,L))
    Walk = Walk.cumsum(axis = 1)
    Walk = Walk.T
    return Walk

R, L = raw_input("Enter #Walks, Length of Walk: ").split(',')
R = int(R)
L = int(L)

fig, ax = plt.subplots()
plt.plot(RandomWalkLine(R,L))
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

def RandomWalk(R, L):
    Walk = np.random.uniform(low = -1, high = 1, size =(R,L))
    Walk = Walk.cumsum(axis = 1)
    Walk = Walk[:,L-1]
    random_histo = [last_spot for last_spot in Walk]
    return random_histo

fig, ax = plt.subplots()
plt.hist(RandomWalk(10000,1000))  #This will be a normal distribution.
plt.show()
```

```python
def DoubleSlit(P, L1, U1, L2, U2):
    RandomWalks = np.random.uniform(low = -1, high = 1, size =(5000,1000))
    WalkDS = RandomWalks.cumsum(axis = 1)
    WalkDS = WalkDS[((WalkDS[:, P] > L1) & (WalkDS[:, P] < U1))| (WalkDS[:, P] > L2) & (WalkDS[:, P] < U2)]
    WalkDS = WalkDS.T
    return WalkDS

fig, ax = plt.subplots()
plt.plot(DoubleSlit(700, -40, -30, 30, 50))
plt.show()
```
