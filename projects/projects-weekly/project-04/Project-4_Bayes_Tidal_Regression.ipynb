### All about that Bayes

The Bayes_Watch class conducts a Bayes regression analysis using the following inputs from the user: log posteriori, observed data, the number of walkers for the Markov Chain Monte Carlo simulation, and a list of variables used in the regression. In its current state, the regression for this project is hardcoded into the Class. In future iterations of this project, additional flexibility will be added into the Bayes_Watch so that a user-defined regression can be passed through the class.

Bayes_Watch allows for the following actions to be taken:
- Runs the Markov Chain Monte Carlo simulation upon instantiation of a class object (100,000 steps per walker).
- Estimates model parameters of our regression.
- Creates a scatter plot of our data as well as our model.
- Plots the probability density of our model parameters.
- Creates a corner plot of our model parameters.
- Creates a predictive distribution of values based on a user-specified time.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad, nquad
from scipy.special import gamma
from math import log
import emcee
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import corner
import collections
%matplotlib inline

#Pre-defined likelihood times prior, to give us our posterior distribution
def lnprob(x):
    if x[3]<0:
        return -np.infty
    else:
        return -0.5*x[3]*np.sum([(e[0] - x[0] - (x[1]*np.sin(e[1]) + x[2]*np.cos(e[1])))**2 for e in data]) \
               + 0.5*len(data)*np.log(x[3]) - 0.5*(x[0]**2+ x[1]**2+ x[2]**2)-x[3]
        
class Bayes_Watch(object):
    """Accepts four arguments: Log of posteriori function, input data, number of walkers, and list of variables."""
    def __init__(self, post, data, walkers, variables):
        self.post = post
        self.data = data
        self.walkers = walkers
        self.variables = variables
        self.values = []

        #Run mcmc
        p0 = np.random.rand(self.walkers*len(self.variables)).reshape((self.walkers,len(self.variables)))*10
        sampler = emcee.EnsembleSampler(self.walkers, len(self.variables), self.post)
        pos, prob, state = sampler.run_mcmc(p0, 1000)
        sampler.reset()
        pos, prob, state = sampler.run_mcmc(pos, 100000)
        self.samples = sampler.flatchain
        
        print ('init and mcmc complete.')
        
    def estimate_model_parameters(self):
        '''Call using () to set values of model parameters'''
        for i in range(len(self.variables)):
            self.values.append(1/float(len(self.samples))*sum([e[i] for e in self.samples]))
        dimension_values = collections.OrderedDict(zip(self.variables,self.values))
        return dimension_values
        
    def plot_scatter(self, reg_plot_range, title = 'Regression over data', x_label = 'x', y_label = 'y'):
        '''Create a scatter plot of data and regression line. reg_plot_range=tuple, ex: (0,9); 
        Title="Regression over data"; x_label="x"; y_label="y"'''
        Xval = np.arange(reg_plot_range[0],reg_plot_range[1],0.01) 
        Reg = [self.values[0] + self.values[1]*np.sin(x) + self.values[2]*np.cos(x) for x in Xval]
        X = [x[0] for x in self.data]
        T = [x[1] for x in self.data]
        fig, ax = plt.subplots(1,1, figsize = (5,3))
        ax.scatter(T,X)
        ax.plot(Xval, Reg, label = title, color = 'red')
        ax.set_xlabel(x_label)
        ax.set_ylabel(y_label)
        ax.set_title(title)
        ax.legend(loc = 'best', fontsize='small')
        ax.grid()

    def plot_probability_density(self):
        '''Plot probability densities of model parameters'''
        sample_data = pd.DataFrame(self.samples, columns =[self.variables[0], self.variables[1], 
                                                           self.variables[2], self.variables[3]])
        sample_data.plot(kind = 'density')

    def __pd_mc__(self,func):
        return 1./len(self.samples)*np.sum([func(x) for x in self.samples])
    
    def predict_dist(self, time, func, title="Predictive Distribution"):
        '''Plots prediction distribution around a user-specified time. Inputs: self; time; function, 
           title="Predictive Distribution".'''
        self.title = title
        self.time = time
        self.func = func
        predictiveDist = lambda x: np.vectorize(lambda t: self.__pd_mc__(lambda d: self.func(t,x)(d)))
        Xval = np.arange(0,25)
        plt.plot(Xval,predictiveDist(self.time)(Xval))
        plt.title(self.title)

    def plot_corner(self):
        '''Call () to plot corner plot'''
        fig = corner.corner(self.samples, labels=self.variables)
```
