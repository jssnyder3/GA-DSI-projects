```python
# Clean interest rate to be a float, percentage.
def clean_int(x):
    if '%' in str(x):
        x = str(x).strip('%')
        return float(x)/100
    else:
        return float(x)

loans_df_a.loc[:,('int_rate')] = loans_df_a.loc[:,('int_rate')].apply(clean_int).astype(np.float64)
loans_df_a.head()
```

```python
# We only want to look at fully paid loans (good!) and default and charged off loans (bad!).
# The remaining loan statuses will be replaced with a null value.
def loan_status_filter(x):
    if x == 'Fully Paid':
        return 'Good'
    elif x == 'Default' or x == 'Charged Off':
        return 'Bad'
    else:
        return np.nan

# new column
loans_df_a['loan_status_str'] = [loan_status_filter(x) for x in loans_df_a['loan_status']]
#loans_df_a = loans_df_a[loans_df_a['status'].isin([0, 1])]

loans_df_a.head()
```

Since joint annual income is reported for some loans, I'd like to use annual income as a default but use joint annual income when it's available. I'll do the same for debt-to-income, as well.

```python
def clean_joint(x, y):
    if y > 0:
        return y
    else:
        return x
loans_df_a['annual_income'] = [clean_joint(x,y) for x,y in 
                               zip(loans_df_a['annual_inc'],loans_df_a['annual_inc_joint'])]
loans_df_a['dti_clean'] = [clean_joint(x,y) for x,y in 
                               zip(loans_df_a['dti'],loans_df_a['dti_joint'])]
loans_df_a.head()
```

```python
# Next, let's remove some of the null values in our data set.
loans_df_a = loans_df_a.dropna(subset=['loan_status_str','int_rate', 'funded_amnt', 'home_ownership', 
                                       'annual_income', 'dti_clean','term'])
```

### Normalizing data for further analysis

In order to put the data into a classification, we need to normalize our variables so that no one variable or set of variables gets inadvertently weighted more than others. I'll do this in the code below and then finalize the dataframe for the classification.

```python
loans_df_a.head()
```

```python
loans_df_b = pd.merge(pd.get_dummies(loans_df_a['term'],drop_first=True),loans_df_a,right_index=True,left_index=True)
loans_df_c = pd.merge(pd.get_dummies(loans_df_a['home_ownership'],drop_first=True),\
                      loans_df_b,right_index=True,left_index=True)
loans_df_d = pd.merge(pd.get_dummies(loans_df_a['loan_status_str'],drop_first=True),\
                      loans_df_c,right_index=True,left_index=True)
loans_df_d.head()
```

```python
# Normalize non-dummy data
# NOTE: We need to normalize all data, but we could also do this without creating an individual
# normalization method for each variable. Normalizing the way I've done below is brute-forcing it
# quite a bit, but gives flexibility later on in the case we want to un-normalize the data. 
#  \_()_/

IR_min_val = loans_df_d['int_rate'].min()
IR_max_val = loans_df_d['int_rate'].max()
def IR_min_max(x):
    return (x-IR_min_val)/(IR_max_val-IR_min_val)
loans_df_d['int_rate_norm'] = loans_df_d['int_rate'].apply(IR_min_max)

AI_min_val = loans_df_d['annual_income'].min()
AI_max_val = loans_df_d['annual_income'].max()
def AI_min_max(x):
    return (x-AI_min_val)/(AI_max_val-AI_min_val)
loans_df_d['annual_income_norm'] = loans_df_d['annual_income'].apply(AI_min_max)

DTI_min_val = loans_df_d['dti_clean'].min()
DTI_max_val = loans_df_d['dti_clean'].max()
def DTI_min_max(x):
    return (x-DTI_min_val)/(DTI_max_val-DTI_min_val)
loans_df_d['dti_norm'] = loans_df_d['dti_clean'].apply(DTI_min_max)

FA_min_val = loans_df_d['funded_amnt'].min()
FA_max_val = loans_df_d['funded_amnt'].max()
def FA_min_max(x):
    return (x-FA_min_val)/(FA_max_val-FA_min_val)
loans_df_d['funded_amnt_norm'] = loans_df_d['funded_amnt'].apply(FA_min_max)
```

# Strike a pose -- it's time to model!
Our analysis is going to use a classification model in order to capture the probability of a particular loan failing or not. This will be a pretty straightforward formula:

$$f(x) = \omega_0 + (\omega_1*RentHome) + (\omega_2*OwnHome) + (\omega_3*60moLoanTerm) + (\omega_4*Income) + (\omega_5*DTI) + (\omega_6*FundedAmount) + (\omega_7*IntRate)$$

In Bayesian statistics, we use a 'prior' distribution and a 'likelihood' distribution in order to get a posterior distribution. Our model will use the Bernoulli Distribution as the likelihood and a Normal Distribution as the prior.

## Likelihood
In order to calculate probability, our output needs to be a value between 0 and 1. Placing the above f(x) into a sigmoid function, $\sigma(f(x)) = \frac{1}{1+e^{-f(x)}}$, forces our probability, p, into a range of 0 to 1. Sigmoid functions also introduce nonlinearity to our regression, which we'll handle later on when predicting probabilities of loan repayment.

We will also add constraints to the likelihood. Namely, if the loan has been successfully paid off, we will return the logarithm of our sigmoid, $log(\sigma(f(x))$. If the loan has _not_ been successfully paid off, we will return $log(1-\sigma(f(x)))$.

## Prior
We'll use a normal distribution for a prior, which, when 'logged', looks like the following:

$$ log(Prior) = -\frac{1}{2}(\omega_0^2 + \omega_1^2+ \omega_2^2 + \omega_3^2 + \omega_4^2 + \omega_5^2 + \omega_6^2 + \omega_7^2) $$

We'll then run all this through an MC sampler in order to determine values for our omegas.

```python
sigmoid = lambda x: 1./(1+np.exp(-x))
```

```python
# log likelihood
def lnlikelihood(o, d):
    reg = o[0] + o[1]*d[1] + o[2]*d[2] + o[3]*d[3] + o[4]*d[4]\
          + o[5]*d[5] + o[6]*d[6] + o[7]*d[7]
    if d[0] == 1:
        return np.log(sigmoid(reg))
    else:
        return np.log(1-sigmoid(reg))

#log prior
def lnprior(a):
    # prior is normal distribution
    return -0.5*np.sum([e**2 for e in a])
    
# log posterior 
def lnprob(a):
    return lnprior(a) + np.sum([lnlikelihood(a, e) for e in loan_data])
```

```python
# Hammer Time: Run MCMC with defined dimensions, walkers, and steps
ndim = 8 # 1 lambda and 7 omegas
nwalkers = 20
first_steps = 1000
second_steps = 10000
      
p0 = np.random.rand(nwalkers*ndim).reshape((nwalkers,ndim))
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob)
pos, prob, state = sampler.run_mcmc(p0, first_steps)
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, second_steps)
samples = sampler.flatchain
```

```python
# Corner plot!
fig = corner.corner(samples, labels=['omega0', 'omega1', 'omega2', 'omega3', 'omega4', 'omega5', 'omega6', 'omega7'])
```

```python
# Calculate probability based on scaled data
def predict_prob(o,d):
    # Return probability
    return sigmoid(o[0] + o[1]*d[1] + o[2]*d[2] + o[3]*d[3] + o[4]*d[4]\
           + o[5]*d[5] + o[6]*d[6] + o[7]*d[7])

def E(f, samples,data):
    return 1./float(len(samples))*np.sum([f(s,data) for s in samples])
```
