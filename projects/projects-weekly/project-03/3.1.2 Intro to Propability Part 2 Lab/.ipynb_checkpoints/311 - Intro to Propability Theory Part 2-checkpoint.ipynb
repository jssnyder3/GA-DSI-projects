## A short Recap of Propability Theory - Part 1

**What did we speak about?**

* We talked about integration with respect to various measures. We showed how the integrals of the function $f(x) = \frac{1}{2} \mathbb{1}_{[-1,1)}$ with respect to the following measures differ

$$ \mu_1(A) = \lambda_L(A) \, \text{ and } \mu_2(A) = \lambda_L(A) + \delta_0(A)$$

Namely:

$$\int_{[-2,1)} d\mu_1(x) f(x) = \frac{1}{2} \lambda_L([-2,1) \cap [-1,1)) = 1$$

and

$$\int_{[-2,1)} d\mu_1(x) f(x) = \frac{1}{2}(\lambda_L([-2,1) \cap [-1,1)) + \delta_0([-2,1) \cap [-1,1)) = 2$$

* We defined stochastic independece: If $A,B$ are two events then $A,B$ are called **independent** if 

$$P(A \cap B) = P(A)*P(B)$$

We used two random variables $X_1$ and $X_2$ representing the outcome of two independent throws of a dice as an example.

$$P(X_1 = 2 \cap X_2 = 4) = P(X_1^{-1}(2) \cap X_2^{-1}(4)) = P(X_1 = 2)P(X_2 = 4)$$

* We talked about the **joint distribution** of two random variables $X,Y$

$$ P(X \in A \cap Y \in B) = \int_{AxB}dP(x,y)$$

* Starting with a joint propability distribution we defined the **marginal distribution**
$$P_X(A) = P(X \in A) = \int_{Ax\mathbb{H}}dP(x,y)$$
which again defined a propability measure. 

We looked at the following example:
$$P(X \in A, Y \in B) = \int_{AxB} d\lambda(x,y) \frac{e^{-0.5(x^2+y^2)}}{2\pi}$$
and we discussed that
$$P_X(A) = P(X \in A, Y \in \mathbb{R}) = \int_{Ax\mathbb{R}} dP(x,y) = \int_{A}d\lambda(x)\frac{e^{-0.5x^2}}{\sqrt{2\pi}}$$

* We defined the *conditional propability*. What is the propability of $A$ under the assumption that we already know that $B$ occured.
$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
We recognized that $A \to P(A|B)$ is again a propability measure and thus defined the **conditional expectation**

$$E[f|B] = \int dP(x|B) f(x)$$

* For two disjoint sets $B_1,B_2$ such that $B_1 \uplus B_2 = \Omega$ the disjoint union is the full event space we derived the **sum rule**
$$P(A) = P(A|B_1)P(B_1) + P(A|B_2)P(B_2)$$

## Moments
Before going deeper into special distributions let's discuss how we can characterize distributions. Let's say we have a random variable $X \sim P_X$ which is distributed like $P_X$. 
We are going to define the **n-th moment $M_n$** of a distribution $P_X$ or equivalently the n-th moment of a random variable $X$ by

$$M_n = E[X^n] = \int dP_X(x) x^n$$

as the expectation value of $X^n$.

```python
X = np.arange(-1,1.01,0.0001)
f = lambda n,x : x**n
fig, ax = plt.subplots(1,2, figsize=(10,2))
ax[0].plot(X,f(1,X), label ='$x$')
ax[1].plot(X,f(2,X), label ='$x^2$')
ax[0].plot(X,f(3,X), label ='$x^3$')
ax[0].plot(X,f(5,X), label ='$x^5$')
ax[1].plot(X,f(4,X), label ='$x^4$')
ax[1].plot(X,f(6,X), label ='$x^6$')
for e in ax:
    e.legend(loc = 'best', fontsize = 'small')
    e.set_xlim([-1,1])
    e.grid()
```

```python
X = np.arange(-5,5.01,0.0001)
f = lambda n,x : x**n
fig, ax = plt.subplots(1,2, figsize=(10,2))
ax[0].plot(X,f(1,X), label ='$x$')
ax[1].plot(X,f(2,X), label ='$x^2$')
ax[0].plot(X,f(3,X), label ='$x^3$')
ax[0].plot(X,f(5,X), label ='$x^5$')
ax[1].plot(X,f(4,X), label ='$x^4$')
ax[1].plot(X,f(6,X), label ='$x^6$')
for e in ax:
    e.legend(loc = 'best', fontsize = 'small')
    e.set_ylim([-100,100])
    e.set_xlim([-5,5])
    e.grid()
```

For increasing n the weight of the function moves further and further out. Even exponents are always positve whereas uneven exponents flip its sign. With repect to a distribution we can infere that even exponents measure the weight of the tails further away from the origin, whereas uneven exponents measure the difference between the tails on either side.

```python
norm = lambda lam, mu: lambda x: lam/(np.sqrt(2*np.pi))*np.exp(-0.5*lam*(x-mu)**2)
m = lambda n: lambda x: norm(1,0.5)(x) * f(n,x)
fig, ax = plt.subplots(1,3, figsize=(10,3))
ax[0].plot(X, m(1)(X), label = 'n = 1')
ax[1].plot(X, m(2)(X), label = 'n = 2')
ax[0].plot(X, m(3)(X), label = 'n = 3')
ax[1].plot(X, m(4)(X), label = 'n = 4')
ax[0].plot(X, m(5)(X), label = 'n = 5')
ax[2].plot(X, norm(1,0.5)(X), label ='Distribution')
for e in ax:
    e.legend(loc = 'best', fontsize = 'small')
    e.grid()
```

The 1st Moment of a distribution is called th **Mean** and is often denoted by $\mu_X$.

$$M_1 = \mu_X = E[X]$$

From our previous discussions we know that $\mu_X = E[X]$ is the weighted avergae value of the random variable $X$. We used a portfolio with various interest rates as an example. Instead of measuring the weight from the origin we can also measure the weights from the weighted average value.

The **Variance** is defined by $$Var[X] = E[(X - E[X])^2] = E[X^2] - E[X]^2 = M_2 - M_1^2$$

We can interpret the Variance as the weighted average disctance squared from the mean. The square root of the variance is often denoted by $\sigma_X = \sqrt{Var[X]}$

In terms of units the mean has a unit of length where as the Variance has a unit of length squared.

Thus we can define **normalized moments** by $$M^{norm}_n = E[(\frac{(X-\mu_X)}{\sigma_X} )^n]$$

**Moment Theorem!! Important and never forget**

Let's say we have two random variables $X,Y$ which are distributed in some way. If all the moments are equal $$E[X^n] = E[Y^n] \text{ for all n}$$ then $X$ and $Y$ have the same distribution!! We can thus compare two distributions by comparing their Moments with each other!!

Problem: Sometimes the moments don't exist. Dont' worry about it though.

## Special Distributions

* **Bernoulli Distribution**

We already worked with the Bernoulli Distribution. The Bernoulli Distribution represents a scenario where there are only two outcomes 1 and 0. The coin flipping experiment was an example of the Bernoulli Distribution.
Let's say we have two events $X=0$ and $X=1$ with propabilities $P(X = 0) = p$ and $P(X=1) = (1-p)$. We can write the Bernoulli Distribution as

$$Bern_p(X=x) = p^x*(1-p)^{1-x}$$

Simple calculation lead to

$$E[X^n] = p \text{ and thus } \mu_X = E[X] = p \text{ and } Var[X] = p - p^2 = p(1-p)$$

And of course $$Bern_p(X = 1 \cup X = 0 ) =1$$

* **Binomial Distribution**

Let's say we have N independent random variables $(X_1,X_2,..,X_N)$ with each one identically distributed with the Bernoulli Distribution $Bern_p$ for a fixed value of p. As in the lab we can now define the sum of all of the random variables $$Y = \sum_{k=1}^{N} X_k$$

$Y$ is then again a random variable and is distributed with the Binomial Distribution defined by

$$Bin_{(N,p)}(Y = k) = \frac{N!}{(N-k)!k!}p^k(1-p)^{N-k}$$

So if we flip a coin 10 times then the propability of flipping 6 heads in total is $$Bin_{(N=10,p=1/2)}(Y=6) = \frac{10!}{4!6!}(1/2)^6(1-1/2)^4 = 0.205..$$

The mean and the variance compute to:

$$E_{(N,p)}[Y] = Np \, \text{ and } Var_{(N,p)}[Y]= Np(1-p)$$

```python
Binom = lambda N,p: lambda k: binomial(N,k)*(p**k)*(1-p)**(N-k)
k = np.arange(0,11,1)
fig, ax = plt.subplots(1,3, figsize=(10,3))
fig.subplots_adjust(wspace = 0.5)
ax[0].bar(k,Binom(10,1/10.)(k), label = 'N = 10, p = 1/10')
ax[1].bar(k,Binom(10,1/4.)(k), label = 'N = 10, p = 1/4')
ax[2].bar(k,Binom(10,0.9)(k), label = 'N = 10, p = 0.9')
for e in ax:
    e.legend(loc ='best', fontsize = 'small')
    e.set_xlabel('k')
    e.set_ylabel('P(Y=k)')
```

**Poisson Distribution**

* Let say we have a set intervall of time or a set area of space. Further let $N$ be a random variable denoting the number of events in that given time intervall or area of space. If we assume that the events occur at a constant rate we can model the distribution of N using the Poisson Distribution

$$Pois_{\lambda}(N = k) = \frac{\lambda^{k} e^{-\lambda}}{k!}$$

An example would be the number of storms in a year or the number of goals in a soccer match. 

$$Pois_{\lambda}(N=0) = e^{-\lambda}$$

Thus $\lambda = -log(P_{\lambda}(N=0))$ is the log of the propability that no event will occure within a certain inetrvall.
The mean and the variance are given by:

$$\mu_N = E[N] = \lambda \text{ and } \sigma_N = \sqrt{Var[N]} = \lambda$$

Thus $\lambda$ also represents the expected number of occurences per intervall.

```python
Pois = lambda lamb: lambda k: lamb**k*np.exp(-lamb)/np.math.factorial(k)
N = np.arange(0,11,1)
PoisNum = lambda lam: np.array([Pois(lam)(n) for n in N])
fig, ax = plt.subplots(1,2, figsize =(7,3))
ax[0].bar(N,PoisNum(0.8), label = '$\lambda = 0.8$')
ax[1].bar(N,PoisNum(4), label = '$\lambda = 4$')
for e in ax:
    e.legend(loc = 'best')
    e.set_xlabel('N = k events')
```

**Beta Distribution**

The Beta distribution is defined for random variables $X:\Omega \to [0,1]$ which take values only between 0 and 1.

$$Beta_{(a,b)}(X \in A) = \int_{A} d\lambda_L(x) \frac{\Gamma(a+b)}{\Gamma(a) \Gamma(b)}x^{a-1}(1-x)^{b-1}$$

Or in other notation

$$dBeta_{(a,b)}(x) = \frac{\Gamma(a+b)}{\Gamma(a) \Gamma(b)}x^{a-1}(1-x)^{b-1} d\lambda_L(x)$$
$\Gamma$ is the gamma function which is an extenstion of the factorial to real numbers. We will talk about examples where the Beta Distribution is used.

```python
Beta = lambda a,b: lambda x: gamma(a+b)/(gamma(a)*gamma(b))*x**(a-1)*(1-x)**(b-1)
X = np.arange(0,1.01,0.01)
a = [4,0.5,0.2,2]
b = [3,3,0.2,0.2]
ziped = zip(a,b)
fig, ax = plt.subplots(2,2,figsize=(10,5))
ax = ax.reshape(4,)
ziped2 = zip(ax,ziped)
for e in ziped2:
    e[0].plot(X,Beta(*e[1])(X), label = 'a = {0}, b={1}'.format(*e[1]))
    e[0].legend(loc = 'best', fontsize = 'small')
```

The *hyperparameter* a determines the shape near $X=0$ where as the *hyperparameter* b determines the shape near $X=1$.

$$\mu_{(a,b)} = E_{(a,b)}[X] = \frac{a}{a+b} \text{ and } \sigma_{(a,b)}^2 = Var_{(a,b)}[X] = \frac{ab}{(a+b)^2(a+b+1)}$$ 

**Normal Distribution**

The normal distribution is defined by

$$dNorm_{(\lambda,\mu)}(x) = \sqrt{\frac{\lambda}{2\pi}} e^{\frac{\lambda}{2}(x-\mu)^2}d\lambda_L(x)$$ 

The *hyperparameter* $\lambda$ is called the precision and and the *hyperparameter* $\mu$ is called the mean.

$$M_1 = E[X] =\mu \text{ and } Var[X] = \lambda^{-1}$$

The Normal Distribution is important because of the **Central Limit Theorem**. Namely let's say we have N random variables $(X_1, X_2, ..., X_N)$ which are all indenpendent and identically distributed with mean $E[X]$ and variance $Var[X]$. They can be distributed in anyway. The random variable $$Y = \sqrt{\frac{N}{Var[X]}} \sum_{k=1}^{N} (X_k-E[X])$$ can be approximated using a normal distribution with $\mu = 0$ and $\lambda = 1$.

```python
Norm = lambda lamb, mu: lambda x: np.sqrt(lamb/(2*np.pi))*np.exp(-lamb/2*(x-mu)**2)
lamba =[20,5,1,0.5]
X = np.arange(-5,5.01,0.001)
fig, ax = plt.subplots(2,2, figsize =(10,5))
ax = ax.reshape(4,)
elements = zip(ax,lamba)
for e in elements:
    e[0].plot(X,Norm(e[1],0)(X), label='$\lambda ={0} , \mu = 0$'.format(e[1]))
    e[0].legend(loc = 'best', fontsize = 'small')
    e[0].set_xlim([-4,4])
    e[0].set_ylim([0,2])
    e[0].grid()
```

Another important feature is the following relationship between the Normal Distribution and the Dirac Measure

$$\lim_{\lambda \to \infty} Norm_{(\lambda,\mu)} = \delta_{\mu}$$

And thus for large precisions or equivalently for small variances we can approximate the expectation value of a function by:

$$E[f] = \int dNorm_{(\lambda,\mu)}(x) f(x)\to \int d\delta_{\mu}(x) f(x) = f(\mu)$$

*Rule of thumb:* Use the Normal Distribution if you don't know anything about a Random Variable other than that it is somehow centered around a value and can be a positive or negative real numebrs.

**Gamma Distribution**

The Gamma distribution is defined by
$$d\Gamma_{(\beta, \alpha)}(x) = \frac{\beta^{\alpha}}{\Gamma(\alpha)} x^{\alpha -1} e^{-\beta x}d\lambda_L(x) ,\, \beta >0, \alpha >0$$

A random Variable which is distributed with the Gamma Distribution only takes positive values. This distribution is often used to model waiting times such as how long do I have to wait at the DMV.
