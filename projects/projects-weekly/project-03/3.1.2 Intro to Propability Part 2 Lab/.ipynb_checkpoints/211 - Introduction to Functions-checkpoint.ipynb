## What is a Function?

Consider two sets $A = \{a_1, a_2,...\}$ and $B = \{b_1, b_2, ...\}$ of any size. A function $F$ from $A$ to $B$ is a mapping that maps certain elements from $A$ into $B$.
$$ F:A \to B $$
- The set $F(A) \subset B$ is called the image of $A$ under $F$.  
- Let $b \in B$ be any element of $B$ then $F^{-1}(b) \subset A$ is the set of all elements in A that are mapped to $b$.
- A function $F$ for which $F^{-1}(b) \subset A$ only consists of one element for every b is called a one - to - one mapping.

***Example 1***

The set of all real numbers is denoted by $\mathbb{R}$. Consider the following mapping:
$$ F:\mathbb{R} \to \mathbb{R};  F(x) = x^2$$
As we already know $x^2$ only takes positive values. Thus the image $F(\mathbb{R})$ is equal to the set of all positive real numbers denoted by $\mathbb{R}^{+}$.
Let $y \in \mathbb{R}^{+}$ be any positive number such as 5 for example. Then:
$$F^{-1}(y) = \{+\sqrt{y}, -\sqrt{y}\}$$
Thus the mapping is not one-to-one.


***Example 2***

Consider the following mapping:
$$ F_{N}:\mathbb{R} \to \mathbb{R};  F_{N}(x) = \sum_{k=0}^{N} \frac{1}{k!}* x^k$$
This function actually represents a class of functions for any integer $N \in \mathbb{N}$. Let's try to find out what the image $F_N(\mathbb{R})$ looks like. It is clear that if we choose a positive $x$ value then $F_N(x) >0$. 

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
X = np.arange(-10,11,1)
F_of_X = np.array([x**2 for x in X])
FN_of_X =  lambda N: np.array([sum([1/float(np.math.factorial(k))*pow(x,k) for k in range(N+1)]) for x in X])
fig = plt.figure(figsize=(6,3))
ax1 = fig.add_subplot(1,2,1)
ax1.plot(X, F_of_X, color = 'red', label = 'F(x) = x^2')
ax1.legend(loc = 'best', fontsize = 'small')
ax1.set_xlabel('x')
ax1.set_ylabel('F(x)')
ax2 = fig.add_subplot(1,2,2)
ax2.plot(X, FN_of_X(2), color = 'red', label = 'N = 2')
ax2.plot(X, FN_of_X(3), color = 'blue', label = 'N = 3')
ax2.plot(X, FN_of_X(4), color = 'orange', label = 'N = 4')
ax2.plot(X, FN_of_X(5), color = 'black', label = 'N = 5')
ax2.set_ylim([-500,500])
ax2.legend(loc = 'best', fontsize = 'small')
```

***Example 3***

Consider the following mapping:
$$\lim_{N \to \infty} F_{N} = Exp:\mathbb{R} \to \mathbb{R};  Exp(x) = \sum_{k=0}^{\infty} \frac{1}{k!}* x^k$$

Exp is called the exponential function and is one of the most fundamental functions in mathematics. 
Basic properties:
1. $Exp(x+y) = Exp(x)*Exp(y)$
2. $\frac{1}{Exp(x)} = Exp(-x)$
3. $ (Exp(x))^y = Exp(x*y)$
4. The expotential function is a one-to-one map from $\mathbb{R} \to \mathbb{R}^{+}$. 
5. The exponentail function increases monotoneously. 
6. $\lim_{x \to -\infty} Exp(x) = 0$
7. $\lim_{x \to \infty} Exp(x) = \infty$

***Example 4***

The inverse map of the exponential function is called logarithm. 
$$ Exp^{-1} = Log:\mathbb{R}^{+} \to \mathbb{R}$$

Basic Properties:
1. $Log(Exp(x)) = Exp(Log(x)) = x$ (inverse property)
2. $Log(x^y) = y*Log(x)$
3. $Log(x*y) = Log(x) + Log(y)$
4. $Log(x/y) = Log(x) - Log(y)$
5. The logarithm is a monotoneously incresing function
6. The logarithm is a one-to-one map from $\mathbb{R}^{+} \to \mathbb{R}$
7. $\lim_{x \to 0} Log(x) = -\infty$
8. $\lim_{x \to \infty} Log(x) = \infty$

```python
X = np.arange(-5,6,0.01)
Exp = np.array([np.exp(x) for x in X])
Log =  np.array([np.log(x) for x in X if x > 0])
fig = plt.figure(figsize=(6,3))
fig.subplots_adjust(wspace = 0.3)
ax1 = fig.add_subplot(1,2,1)
ax1.plot(X, Exp, color = 'red', label = 'Exp(x)')
ax1.legend(loc = 'best', fontsize = 'small')
ax1.set_xlabel('x')
ax1.set_ylabel('Exp(x)')
ax1.grid()
ax2 = fig.add_subplot(1,2,2)
ax2.plot([x for x in X if x > 0], Log, color = 'red', label = 'Log(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('Log(x)')
ax2.legend(loc = 'best', fontsize = 'small')
ax2.grid()
```

***Example 5***

Consider the following mapping:
$$ F:\mathbb{R}^2 \to \mathbb{R}^{+}; F(x,y) = Exp(-x^2)*Exp(-y^2)$$

This function maps the 2-dimensional space $\mathbb{R}^2$ onto the 1-dimensional space $\mathbb{R}^{+}$. This map is no longer a one-to-one map! Let's try to understand how $F^{-1}(z) \subset \mathbb{R}^2, z \in \mathbb{R}^{+}$ looks like.

$$ F(x,y) = Exp(-x^2)* Exp(-y^2) = Exp(-x^2-y^2)$$

So if we choose any point in $(x_0,y_0) \in \mathbb{R}^2$ such that $-x^2_0-y^2_0 = Log(z)$ then 

$$F(x_0,y_0) = Exp(-x^2_0 - y^2_0) = Exp(Log(z)) = z$$

We can rewrite the condition to $y^2_0  + x^2_0 = -Log(z)$ and we can see that $F(x,y)$ maps circles of Radius R onto the same value in $\mathbb{R}^{+}$.

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.linspace(-2,2, 100)
Y = np.linspace(-2,2, 100)
X, Y = np.meshgrid(X,Y)
Z = np.exp(-X**2 - Y**2)
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1, projection = '3d')
ax1.plot_surface(X, Y, Z)
ax2 = fig.add_subplot(1,2,2)
ax2.contour(X, Y, Z)
```
